Import and Read Data

```python
import pandas as pd
import numpy as np
df=pd.read_csv(r'https://github.com/Robmattles/Python-For-Data-Analysis/blob/master/Data%20Files/ISIStweetsEnhanced.csv?raw=true')
```

Sorting

```python
dfSorted=df.sort_values(by=['name','time'],ascending=[False,True])
dfSorted
```

Changing Data Types

```python
df['followersstr']=df['followers'].astype(str)
```

Merging

```python
conus=pd.DataFrame({'CONUS Status':['CONUS' if x>.5 else 'OCONUS' for x in np.random.random(20)],
           'Username':list(df['username'].unique())[-20:]})
conus
```

```python
df.merge(conus, left_on='username', right_on='Username',how='left')
```

Adding Column

```python
df['Placeholder']='LOREM IPSUM'
df.head()
```

Concatenating

```python
additionalRows=df.sample(20)
additionalRows
pd.concat([df,additionalRows],axis=0)
```

Melting

```python
hashtagColumns=['Hashtag '+str(x) for x in range(7)]
print(hashtagColumns)
melted=df.melt(id_vars=['name','username','tweets', 'Unnamed: 0'],value_vars=hashtagColumns)
melted[melted['value'].notnull()]
```

Pivoting

```python
list(df.columns)
```

```python
melted.pivot(columns='variable',values='value', index='Unnamed: 0').tail(5)
```

Apply

```python
def addOne(s):
    return s+1
df['numberstatuses'].apply(addOne)
```

Removing Duplicates

```python
df.drop_duplicates(subset=['name','description'])
```
