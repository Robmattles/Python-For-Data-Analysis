# Module 8 - Dates and Time

Import and Read Data

```python
import pandas as pd
import numpy as np
df=pd.read_csv(r'C:\Users\mwbarr\Python for Data Analysis\Data Files\isistweetsenhanced.csv')
```

Brief Exploratory Analysis - Notice Dates stored as non-dates

```python
df.info()
```

Converting to Datetime

```python
df['time']=pd.to_datetime(df['time'])
df['time']
```

Using dt.floor (and mention other functions)

```python
df['time'].dt.floor(freq='d')
df['time'].dt.hour
df['time'].dt.weekday
```

Create your own timedelta

```python
pd.Timedelta(value=5,unit='m')
```

Add time delta to datetime

```python
df['time']+pd.Timedelta(value=5,unit='m')
```

Create time delta from subtracting date time columns

```python
df['New Year']='1/1/2017'
df['New Year']=pd.to_datetime(df['New Year'])
df['length']=df['New Year']-df['time']
df['length']
```

Create time delta by subtracting single date from time column

```python
import datetime
df['time']-pd.to_datetime('1/1/2017')
```

Extracting information out of time deltas using total_seconds

```python
df['length'].dt.total_seconds()
```

Subsetting with time deltas

```python
df[df['length']<pd.Timedelta(value=300,unit='d')]
```

Convert Date time to strings

```python
df['time'].dt.strftime('%d-%m/%Y')
```

Grouping using datetimes

```python
df.groupby(pd.Grouper(freq='m',key='time')).size()
```
