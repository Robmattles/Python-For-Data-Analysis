Grouping on a Single Column

```python
grouped=df.groupby('username')
grouped
```

Simple Aggregation with sum/size

```python
df.groupby('username').size()
```

```python
grouped.sum()
grouped.mean()
grouped.max()
```

Grouping on multiple columns

```python
grouped=df.groupby(['username','location'])
grouped.size()
```

Using a Custom Aggregation Function

```python
def addOneAndSum(l):
    addedOne=l+1
    return l.sum()
grouped.agg(addOneAndSum)
```

Using Multiple functions

```python
grouped.agg({'numberstatuses':np.mean, 'followers':sum})
```

Converting the results of an aggregation back to a dataframe using reset_index and unstack

```python
grouped.sum()['numberstatuses'].reset_index()
```

```python
grouped.sum()['numberstatuses'].unstack()
```

Transforming Data by Group

```python
def subtractMean(l):
    return l-np.mean(l)

grouped=df.groupby('username')
grouped['followers'].transform(subtractMean)
```
