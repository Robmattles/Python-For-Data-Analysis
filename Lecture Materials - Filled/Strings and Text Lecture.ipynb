Regex (Description, Testers, pre-made patterns, then below examples)

Contains - Now with more regex

```python
df[df['username'].str.contains('\d')]['username'].drop_duplicates().head()
```

```python
df[df['username'].str.contains('a\d')]['username'].drop_duplicates()
```

```python
df[df['username'].str.lower().str.contains('a\d{2}')]['username'].drop_duplicates()
```

```python
df[df['username'].str.lower().str.contains('[a-z]\d[a-z]')]['username'].drop_duplicates()
```

Extract

```python
df['tweets'].str.extract('@(\w*)\s').value_counts().head()
```

```python
df['tweets'].str.extract('(http://\S*)',).head()
```

Extract All

```python
df['tweets'].str.extractall('@(\w*)\s')
```

Resetting index on extract all results

```python
df['tweets'].str.extractall('@(\w*)\s').reset_index()
```
